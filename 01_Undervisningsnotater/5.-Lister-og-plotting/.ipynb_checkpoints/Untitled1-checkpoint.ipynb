# 5.1.3 - Matplotlib

### Importerer biblioteket:

```python
import matplotlib.pyplot as plt
plt.style.use('dark_background')
%config InlineBackend.figure_format = 'svg'
```

Lister med data vi ønkser å plotte

```python
x = [0, 1, 2, 3, 4, 5]
y = [0,  1, 4, 9, 16, 25]
```

Enkelt plot uten info om grafen

```python
fig = plt.plot(x,y)
```

Legge til plot-informasjon

```python
plt.xlabel('$x$') # Merker x-aksen
plt.ylabel('$f(x)$') # Merker y-aksen
plt.plot(x, y, '-.', label='$f(x)=x^2$')
plt.show()
```
