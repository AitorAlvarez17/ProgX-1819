## Grunnleggende om matplotlib

Matplotlib er et bibliotek, og må importeres før det kan brukes. 

Hvis du bruker Dark theme

```python
plt.style.use('dark_background')
```

Importerer biblioteket og setter formatet til svg

```python
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'
```

Det trenger to (eller flere) like lange lister med punkter for å kunne plotte. Listene danner til sammen en rekke med punkter.

Listene over danner punktene (0,0), (1,1), (2,4), (3,9) og (4,16) som hver og en blir plottet og forbundet med en rett linje.

```python
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]
```

Lage plottet

```python
fig = plt.plot(x, y)
```

Som regel ønsker vi mer informasjon i plottet. I eksempelet under ser vi hvordan vi kan legge til mer informasjon.

```python
plt.xlabel('$x$') # Merker x-aksen
plt.ylabel('$f(x)$') # Merker y-aksen
plt.title('En tittel') #Tittel
plt.plot(x, y, '-.', label='$f(x)=x^2$')
plt.legend() # Infoboksen øverst i hjørnet. Bruker data fra label
plt.show()
```

## Numpy og funksjoner i matplotlib

```python
plt.plot(np.sin(np.linspace(0, 20, 100)))
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)


plt.figure(figsize=(15,8))
plt.plot(x, y,  label='$sin(x)$')

# dashes=[lengde, mellomrom]
plt.plot(x, y - 0.2, dashes=[2, 6], label='$sin(x)-0.2$')

plt.legend()
plt.show()
```
