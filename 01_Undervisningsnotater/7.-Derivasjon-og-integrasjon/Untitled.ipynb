# 7.1 Derivasjon

## Definisjonen av den deriverte

Den deriverte til en funksjon $f$ i et punkt $a$ er stigningstallet mellom a og et punkt som er uendelig nærme a. Vi kaller dette punktet $a+\Delta x$, og den deriverte er gitt ved:

$$f^\prime(a)=\lim_{\Delta x \to 0} \frac{f(a+\Delta x) - f(a)}{\Delta x}$$

## Newtons kvotient
Vi tilnærmer den deriverte med Newtons kvotient. Den deriverte til en funksjon $f$ kan tilnærmes slik dersom $h$ er liten.

$$f^\prime(a)=\lim_{\Delta x \to 0} \frac{f(a + \Delta x) - f(a)}{\Delta x} \approx \frac{f(a + h) - f(a)}{h}$$


## Implimentering i Python

```python
def derivert(f, a, h=10**-8):
    return (f(a + h) - f(a))/h
```

### Eksempel 1

```python
def f(x):
    return x**2
```

```python
print(f'{derivert(f,2):.7f}')
```

### Eksempel 2

$$g(x)=x^4 - 1$$

$$g^\prime (3)= 4*3^3=108$$
Plot diferansen mellom den eksakt deriverte til $g^\prime (3)$ og den tilnærmede numeriske med ulike $h$.


```python
def g(x):
    return x**4 - 1
```

```python
import numpy as np
```

```python
differ = []
for i in range(3,14):
    deriverte = derivert(g, 3, 10**-i)
    diff = np.abs(deriverte - 108)
    differ.append(diff)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(differ)
```
