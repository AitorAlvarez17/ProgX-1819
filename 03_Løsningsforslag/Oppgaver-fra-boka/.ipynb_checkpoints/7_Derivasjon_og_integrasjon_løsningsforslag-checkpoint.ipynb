### Oppgave 7.0

Newtons kvotient

#### Løsning:

```python
def derivert(f, x, delta_x=1e-08):
    '''
        Deriverte med Newtons kvotient
        Input: f: funksjonen, x: x-verdi, delta_x: punktavstand
    '''
    return ( f(x+delta_x) -  f(x) )/delta_x 
    
```

-----------

```python
def utskrift(f, navn):
    print(f'{navn}´(2) = {derivert(f,2):.1f}')
    print(f'{navn}´(3) = {derivert(f,3):.1f}')
    print(f'{navn}´(4) = {derivert(f,4):.1f}')
```

```python
def graf(f):
    f_verdier = np.zeros(len(x_er))
    f_deriverte_verdier = np.zeros(len(x_er))

    for x in range(len(x_er)):
        f_verdier[x] = f(x)
        f_deriverte_verdier[x] = derivert(f, x)

    plt.plot(x_er, f_verdier, label='f')
    plt.plot(x_er, f_deriverte_verdier, label='f-deriverte')
    plt.legend()
    plt.title('Funksjonen og den deriverte', fontsize=16)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show() 
```

```python
x_er = np.linspace(2,5,100) # 100 x-verdier fra 2 til 5
```

### Oppgave 7.1

Deriver følgende funksjoner med Newtons kvotient. Plot funksjonene for $x \in [2,5\rangle$. Kontroller ved å derivere for hånd.

----------
**1.**  $\quad f_1(x)=2x+1$
#### Løsning:

```python
def f_1(x):
    return 2*x + 1

utskrift(f_1, 'f_1')
graf(f_1)
```

----------
**2.** $\quad f_2(x)=x^2-4x+5$
#### Løsning:

```python
def f_2(x):
    return x**2 -4*x + 5

utskrift(f_2, 'f_2')
graf(f_2)
```

-----------
**3.**  $\quad f_3(x)=e^x$
#### Løsning:

```python
def f_3(x):
    return np.e**x

utskrift(f_3, 'f_3')
graf(f_3)
```

-----------
**4.**  $\quad f_4(x)=e^x-5x$
#### Løsning:

```python
def f_4(x):
    return np.e**x - 5*x

utskrift(f_4, 'f_4')
graf(f_4)
```

-----------
**5.**  $\quad f_5(x)=6\sqrt x-x$
#### Løsning:

```python
def f_5(x):
    return 6*np.sqrt(x)-x

utskrift(f_5, 'f_5')
graf(f_5)
```

-----------
**6.** $\quad f_6(x)=5e^{-2x}$
#### Løsning:

```python
def f_6(x):
    return 5*np.e**(-2*x)

utskrift(f_6, 'f_6')
graf(f_6)
```

-----------
**7.** $\quad f_7(x)=\sqrt{\ln(x)}$
#### Løsning:

```python
def f_7(x):
    return np.sqrt(np.log(x))

utskrift(f_7, 'f_7')
graf(f_7)
```

-----------
**8.** $\quad f_8(x)=4\ln(x^2)$
#### Løsning:

```python
def f_8(x):
    return 4*np.log(x**2)

utskrift(f_8, 'f_8')
graf(f_8)
```

-----------
**9.** $\quad f_9(x)=4 \cdot 5^{x^3-2x}$
#### Løsning:

```python
def f_9(x):
    return 4*5**(x**3 - 2*x)

utskrift(f_9, 'f_9')
graf(f_9)
```
